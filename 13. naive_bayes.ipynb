## 나이브베이즈 모델 적용(Laplace 적용x)

MultinomialNB,,

먼저 parameters을 통해 모델을 만들고,
그 이후 attributes 을 통해 모델 뒤 함수 . 찍어서 뒤에 확인하는 것
https://scikit-learn.org/stable/modules/naive_bayes.html#multinomial-naive-bayes


```python
from sklearn.naive_bayes import MultinomialNB

# alpha : laplace 추정기
# class_prior : 사전확률 직접 정의
# fit_prior : 사전확률 적합

model = MultinomialNB(alpha=0)
model.fit(data.iloc[:, 1:], data['장르']) 
## 모델 학습시켜주기
## 처음은 독립변수, 뒤에는 종속변수
## iloc -> 행 : 이니 행을 다 불러오고, 열은 1부터 불러오기
## dataframe 안의 자료 복습해오기
```

class_prior 은 사전확률, 구분하고 싶은 최종,, 즉 코미디인지 액션인지.. 


학습시킬 때는 코미디인지 액션인지 다 알고 있음. 2/5 => 코미디일 확률 0.4.. 즉 이런 확률을 알고있는게 사전확률 (훈련시킬 때)


베이즈 정리 보면 B가 나왔을 때 A(코미디/액션)일 확률 -> 사후확률...? ,, p(A) = 사전확률

alpha 값만 잘 적용해주면 됨.

1) alpha=0 : 라플라스 적용하지 않았을 때

2) 라플라스 적용했을 때 

```python
# alpha : alpha값 확인
model.alpha
```

```python
# classes_ : 클래스 이름
model.classes_
```

```python
# class_count_ : 클래스 개수 반환
model.class_count_
## 앞에게 액션이었고, 클래스가 두개 있다는 뜻
```

```python
# feature_count_ : 각 클래스별 단어 개수 등장 횟수
## 연인같은 경우 액션은 0, 코미디는 2, 이런 개수를 세주는것
model.feature_count_
## 위에껀 액션, 아래는 코미디
## 퀴즈를 위해 여기를 잘 공부해보셈
```

```python
# class_log_prior_ : class별 log를 씌운 사전확률
model.class_log_prior_
```

```python
## 수학적인 것 안에 np가 굉장히 많은 것을 차지하고 있음.
np.exp(model.class_log_prior_)
```

날으는을 보면 액션1, 코미디1 임.

사후확률로 보면 P(날으는/액션) = 1/3 (액션 3중에 날으는은 한번 들어감)

이걸 log, 즉 지수승 씌워주면 확률 구할 수 있음

```python
# feature_log_prob_ : 단어별 log를 씌운 사후확률
np.exp(model.feature_log_prob_)
```

## train test 분리

```python
import random

# random seed 설정하기

```

x_data 는 독립변수 -> V1 ~ V16 까지 일것임.

```python
x_data = data.iloc[:,1:] ## iloc 는 다 불러오는 것
y_data = data['Class']
```

```python
from sklearn.model_selection import train_test_split

# train, test set분리 
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, 
                                                    test_size = 0.3, 
                                                    random_state = 7)
## random_state 는 무작위값. 100개 중 random 으로 70,30개 뽑는데, seed 값에 따라 7, 8 뽑는게 다 다름..
## random_state 여러개 해서 평균값 하는 경우도 많음( 한쪽에 치중되게 되었을 수도 있으니 )
```

## 나이브베이즈 모델 적용

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB(alpha = 1)

model.fit(X_train, y_train)

## 모델 훈련이 중요한게 아니라 데이터를 가지고 결측치를 어떻게 처리하는지가 중요함.
## 모델 훈련은 그냥 코드만 돌리면되어서 쉬움. 
## 결측치 하는데 90퍼센트를 써야함.
```

```python
# test 데이터로 예측
y_pred = model.predict(X_test)
```

```python
model.predict(X_test)
```

```python
# 정확도 확인
sum(y_pred == y_test) / len(y_test)
```
