## 1. HouseVotes84데이터에서 NaN의 값을 채워주세요
### 각 행마다 더 많은 y or n의 개수로 NaN의 값을 채워주세요
### 1번째 행은 y가 9개 n은 6개 이므로 NaN의 값을 y로 채우기

```python
data = pd.read_csv("C:/Users/user/Desktop/이해영/큐시즘/머신러닝스터디/python 데이터 분석/data/HouseVotes84.csv")
```

```python
data['n_count'] = (data[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16']] == 'n').sum(axis=1)
data['y_count'] = (data[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16']] == 'y').sum(axis=1)

data
```

```python
for i in range(0, 434) :
    if data['n_count'][i] >= data['y_count'][i] :
        data.iloc[i] = data.iloc[i].fillna('n')
    elif data['n_count'][i] <= data['y_count'][i] :
        data.iloc[i] = data.iloc[i].fillna('y')
        
data
```

## 2. 위에서 NaN값을 채운 data로 train/test를 7:3으로 분리하고 alpha=1을 적용해서 나이브베이즈 모델을 사용한 정확도를 출력해주세요

```python
def convert(x) :
    if x == 'y' :
        return 1
    else : 
        return 0
```

```python
for i in data.columns[1:]:
    data[i] = data[i].apply(convert)

data
```

```python
import random
x_data = data.iloc[:,1:]
y_data = data['Class']

from sklearn.model_selection import train_test_split

# train, test set분리 
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, 
                                                    test_size = 0.3, 
                                                    random_state = 7)

from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB(alpha = 1)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
sum(y_pred == y_test) / len(y_test)
```

## 각 column의 entropy 정보이득을 계산해주는 함수를 만들어주세요

```python
def both(column) : 
    count = 0

    for i in df.index :
        standard = (df[column][i] == 1 and df['y'][i] == 1)
    
        if standard == 1 :
            count += 1
        
    return count
            

def both_no(column) :
    count = 0
    
    for i in df.index :
        standard = (df[column][i] == 0 and df['y'][i] == 0)
    
        if standard == 1 :
            count += 1
        
    return count
```

```python
def entropy(column, root):
    
    if ( both(column) / sum(df[column] == 1) ) != 0 :
        LOG1 = ( both(column) / sum(df[column] == 1) ) * np.log2(( both(column) / sum(df[column] == 1) ))
    else :
        LOG1 = 0
    
    if (1 - ( both(column) / sum(df[column] == 1) )) != 0 :
        LOG2 = (1 - ( both(column) / sum(df[column] == 1) )) * np.log2((1 - ( both(column) / sum(df[column] == 1) )))
    else :
        LOG2 = 0
    
    LOG = LOG1 + LOG2
    FRONT = (sum(df[column] == 1)) / len(column)
    RE = - FRONT  * LOG
    
    front = 1 - FRONT
    
    
    
    try : 
        if (both_no(column) / (len(column) - (sum(df[column] == 1)))) != 0 :
            log1 = (both_no(column) / (len(column) - (sum(df[column] == 1))))  *  np.log2((both_no(column) / (len(column) - (sum(df[column] == 1)))))
        else :
            log1 = 0
    except ZeroDivisionError:
        log1 = 0
    
    try :
        if (1-(both_no(column) / (len(column) - (sum(df[column] == 1))))) != 0 :
            log2 = (1-(both_no(column) / (len(column) - (sum(df[column] == 1))))) * np.log2((1-(both_no(column) / (len(column) - (sum(df[column] == 1))))))
        else :
            log2 = 0
    except ZeroDivisionError:
        log2 = 0
    
    log = log1 + log2
    re = - front * log
    
    result = root - (RE + re)
    

    print(log1)
    print( (len(column) - (sum(df[column] == 1))))
    print((both_no(column) / (len(column) - (sum(df[column] == 1)))) )
    print(np.log2((both_no(column) / (len(column) - (sum(df[column] == 1))))))
    
    return result
```
